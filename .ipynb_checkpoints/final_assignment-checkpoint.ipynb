```python
# reading data
import pandas
import numpy as np
features = pandas.read_csv('train_bnp.csv',index_col='ID')
features.head()
```

```python
features['v91'].value_counts()
```

```python
#found NaN in data
a=features.count()
b=a[a>65000]
```

```python
#features with NaN
features=features[b.index]
features.head()
```

```python
length=len(features)
k=features.dtypes
float_feat=k[k=='float']
int_feat=k[k=='int64']
# target variable
y=np.array(features['target'])
```

```python
float_feat=float_feat.append(int_feat)
float_feat=float_feat[float_feat.index!='target']
```

```python
features=features[float_feat.index]
features
```

```python
# fill NaN by 0
features.fillna(value=0,inplace=True)
```

```python
# features
x=np.array(features)
```

# Gradient boosting

```python
# cross-validation
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import cross_validation
from sklearn.cross_validation import KFold
import time
import datetime
start_time = datetime.datetime.now()
kf=KFold(length,n_folds=5,shuffle=True)
n=[10]
t=[0]
for i,v in enumerate(n):
    start_time = datetime.datetime.now()
    clf = GradientBoostingClassifier(n_estimators=v,verbose=True,random_state=241,learning_rate=0.3)
    sc=cross_validation.cross_val_score(clf,x,y,scoring='roc_auc',cv=kf)
    t[i]=sc.mean()
    print 'Time elapsed:', datetime.datetime.now() - start_time
```

```python
t
```
